```python
file_path = r"E:\Ind-Aus TDA\Export_dataset_2016-22.xlsx"
df = pd.read_excel(file_path)
```

```python
df.head()
```

```python
reshaped_df = pd.melt(df,
                      id_vars=['Product code', 'Product label'],
                      var_name='Year',
                      value_name='Export Value')

# Convert the 'Year' column to an integer type
reshaped_df['Year'] = reshaped_df['Year'].str.extract(r'(\d+)', expand=False).astype(int)

reshaped_df.head(10)
```

```python
# Define the year for forecasting
forecast_year = 2023

# Create an empty list to store the forecasted values
forecasts = []

# Group data by 'Product code' and 'Product label' to fit a linear regression model for each product
for (product_code, product_label), group in reshaped_df.groupby(['Product code', 'Product label']):
    # Prepare the data for regression
    X = group[['Year']].values  # Independent variable (Year)
    y = group['Export Value'].values  # Dependent variable (Export Value)
    
    # Initialize and train the Linear Regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Forecast the export value for the given year (2023)
    future_X = np.array([[forecast_year]])
    forecast_value = model.predict(future_X)[0]  # Extract the predicted value
    
    # Append the forecast result to the list
    forecasts.append({
        'Product code': product_code,
        'Product label': product_label,
        'Year': forecast_year,
        'Forecasted Export Value': forecast_value
    })

# Convert the forecast results into a DataFrame
forecast_df = pd.DataFrame(forecasts)

forecast_df.head(10)
```

```python
# Specify the file path to save the forecasted data
output_path = "E:\Ind-Aus TDA\Forecasted_2023.xlsx"

# Save the forecasted DataFrame to the specified Excel file
forecast_df.to_excel(output_path, index=False)

# Print a confirmation message
print(f"Forecasted data for 2023 has been successfully saved to: {output_path}")
```
